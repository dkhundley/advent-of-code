## Part 1

```python
# Loading in the crab positions
with open('aoc-day7.txt') as f:
    crabs = list(map(int, (f.readlines()[0].replace('\n', '').split(','))))
```

```python
# Instantiating a placeholder that will hold the most efficient fuel value
most_efficient_fuel = 100000000000000000000000000000000000
```

```python
# Iterating through all potential landing positions based on largest value
for position in range(min(crabs), (max(crabs) + 1)):
    
    # Instantiating a variable to hold the fuel required this round
    fuel = 0
    
    # Iterating through all the crab positions
    for crab in crabs:
        
        # Adding required fuel to total
        fuel += abs(crab - position)
        
    # Checking to see if fuel this round was more efficient than previous rounds
    if fuel < most_efficient_fuel:
        most_efficient_fuel = fuel
```

```python
# Printing the final result
print(f'Most efficient fuel: {most_efficient_fuel}')
```

## Part 2

```python
# Loading in the crab positions
with open('aoc-day7.txt') as f:
    crabs = list(map(int, (f.readlines()[0].replace('\n', '').split(','))))
```

```python
# Instantiating a placeholder that will hold the most efficient fuel value
most_efficient_fuel = 100000000000000000000000000000000000
```

```python
# Iterating through all potential landing positions based on largest value
for position in range(min(crabs), (max(crabs) + 1)):
    
    # Instantiating a variable to hold the fuel required this round
    fuel = 0
    
    # Iterating through all the crab positions
    for crab in crabs:
        
        # Noting how many positions the crab will have to travel
        travel = abs(crab - position)
        
        # Instantiating the starting fuel rate value as 1
        fuel_rate = 1
        
        # Incrementing the fuel required per number of steps the crab must travel
        for step in range(travel):
            fuel += fuel_rate
            fuel_rate += 1
            
        # Continuing on if fuel exceeds most efficient value
        if fuel > most_efficient_fuel:
            continue
    
    # Checking to see if fuel this round was more efficient than previous rounds
    if fuel < most_efficient_fuel:
        print(f'New records efficiency: {fuel}')
        most_efficient_fuel = fuel
```

```python
# Printing the final result
print(f'Most efficient fuel: {most_efficient_fuel}')
```
