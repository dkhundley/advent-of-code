# An Introduction to Heaps
In this notebook, we'll cover a special data structure called a **heap**. Heaps are used to manage queues very efficiently. Why is this the case? Consider a standard list of numbers:

`[ 6, 1, 7, 0, 4 ]`

If we wanted to insert a `2` between the `6` and the `1`, we would have to first slide down all the numbers `1` and after to make room, like this:

`[ 6, _, 1, 7, 0, 4 ]`

And then we can insert our new number:

`[ 6, 2, 1, 7, 0, 4 ]`

If you consider the number of computer operations it would take to do this, it would be one for every time we slide number down the line and then one final operation to insert the new number. Because the number of times we need to slide a number over in our list can change, let's represent that by variable `n`. So, in the computer science community, this would be represented as the following:

$$O(n + 1)$$

If you've not seen this before, let's break down this notation. The "O" represents what we in the computer science commmunity call **Big O** notation. Because computers are ultimately reliant on physical entities (aka CPUs) for its computation, a computer is bound by however quickly it can process **floating point operations**, or **FLOPs**. While modern computers can certainly process a lot of FLOPs, you'd be astounded how much computation is required for certain things. Consider ChatGPT. The training of that model required an insane amount of FLOPs!

So computational efficiency is still extremely important, and one of the simplest implementations of computational efficiency in action is through a heap, more specifcally a **binary heap**. The reason it is binary is because each node has no more than two children. The two implementations of a binary heap are the **max heap** and **min heap**, and the only difference between the two is that one is basically ascending and the other is descending. Going back to our example from before...

`[ 6, 2, 1, 7, 0, 4 ]`

Let's visually represent this as a max heap.
```
       7
     /   \
    4     6
   / \   / 
  0   1 2  
```
The Big O of an implementation like this follows what we refer to as **logarithmic growth**. Logarithms are often referred to as inverses of an exponent, and while that is technically true, it provides no intuition for its importance. Let's step back to something you probably do know: **exponential growth**. You've likely been exposed to the idea of exponential growth where something like a virus (perhaps COVID 19) would spread faster and faster as time passed. If you were to draw this exponential curve on a graph, it might look like the following:

````python
# Generating x values from a small positive number
x = np.linspace(1, 10, 100)

# Calculating the exponential values from the x inputs
y = np.exp(x)

# Plotting the logarithmic function using the data from above
plt.figure(figsize = (8, 6))
plt.plot(x, y, label = 'y = x^2')

# Making it pretty-ish
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show();
````

Notice that the growth rate (or **slope**) of the graph starts off a bit slow and then really starts climbing at around 8 in this graph. Logarithmic growth is indeed the inverse, and looking at how this visually contrasts in this next chart will make the intuition instantly clear:

````python
# Generating x values from a small positive number
x = np.linspace(1, 10, 100)

# Calculating the logarithmic values from the x inputs
y = np.log(x)

# Plotting the logarithmic function using the data from above
plt.figure(figsize = (8, 6))
plt.plot(x, y, label = 'y = log(x)')

# Making it pretty-ish
plt.title('Logarithmic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show();
````

Contrast this chart to the exponential chart we showed before. In the exponential chart, growth (re: slope) started very slowly and then drastically increased faster and faster as time went on. Logarithmic growth is the opposite: growth starts off rapidly but as time goes on, the growth gets slower and almost levels out on the graph.

When it comes to computational efficiency, we want to aim for the lowest number of FLOPs we have to perform. Remember, the Big O of adding / removing items from a typical array or list is:

$$O(n + 1)$$

Let's graph this on a chart.

````python
# Generating x values from a small positive number
x = np.linspace(1, 10, 100)
y = x
y += 1

# Plotting the function using the data from above
plt.figure(figsize = (8, 6))
plt.plot(x, y, label = 'y = x + 1')

# Making it pretty-ish
plt.title('Big O for Adding / Removing Items from Typical List')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show();
````

Naturally, this is a straight line upward, what we in the math community would refer to as **linear**. Fortunately no exponential growth, but there's also no logarithmic growth. To really drive home how much better it would be to have logarithmic growth, let's put both lines on a single chart:

````python
# Generating x values from a small positive number
x1 = np.linspace(1, 10, 100)
y1 = x
y1 += 1

# Plotting the function using the data from above
plt.figure(figsize = (8, 6))
plt.plot(x1, y1, label = 'y = x + 1')

# Generating x values from a small positive number
x = np.linspace(1, 10, 100)

# Calculating the logarithmic values from the x inputs
y = np.log(x)

# Plotting the logarithmic function using the data from above
plt.plot(x, y, label = 'y = log(x)')

# Making it pretty-ish
plt.title('Contrasting the Big Os')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show();
````

Viewing it like this, it's a no brainer: the logarithimc growth is strongly preferable as it helps keep the computational efficency at a low growth rate. And as you might expect, the binary heap follows a logarithmic growth rate. Let's formalize that with this Big O notation

$$O(log(n))$$
