Each monkey has several attributes:

Starting items lists your worry level for each item the monkey is currently holding in the order they will be inspected.
Operation shows how your worry level changes as that monkey inspects an item. (An operation like new = old * 5 means that your worry level after the monkey inspected the item is five times whatever your worry level was before inspection.)
Test shows how the monkey uses your worry level to decide where to throw an item next.
If true shows what happens with an item if the Test was true.
If false shows what happens with an item if the Test was false.
After each monkey inspects an item but before it tests your worry level, your relief that the monkey's inspection didn't damage the item causes your worry level to be divided by three and rounded down to the nearest integer.

The monkeys take turns inspecting and throwing items. On a single monkey's turn, it inspects and throws all of the items it is holding one at a time and in the order listed. Monkey 0 goes first, then monkey 1, and so on until each monkey has had one turn. The process of each monkey taking a single turn is called a round.

When a monkey throws an item to another monkey, the item goes on the end of the recipient monkey's list. A monkey that starts a round with no items could end up inspecting and throwing many items by the time its turn comes around. If a monkey is holding no items at the start of its turn, its turn ends.

In this example, the two most active monkeys inspected items 101 and 105 times. The level of monkey business in this situation can be found by multiplying these together: 10605.

Figure out which monkeys to chase by counting how many items they inspect over 20 rounds. What is the level of monkey business after 20 rounds of stuff-slinging simian shenanigans?

````python
import math

def parse_monkey_info(monkey_info):
    '''
    Parses the monkey information appropriately
    
    Inputs:
        - monkey_info (list): A list of unparsed strings containing the monkey information
        
    Returns:
        - monkey_number (int): The number associated to the monkey
        - monkey_details (dict): The specific details associated to the monkey in a parsed form
    '''
    
    # Iterating of each piece of information
    for line in monkey_info:
        
        # Getting the monkey's number
        if line.startswith('Monkey'):
            monkey_number = int(line.split(':')[0][-1])
            
        # Getting the starting items
        elif line.lstrip().startswith('Starting items:'):
            currently_holding = [int(item) for item in line.split(':')[1].replace(' ', '').split(',')]
            
        # Getting the operations
        elif line.lstrip().startswith('Operation:'):
            operations = line.split('=')[1][5:].split(' ')
            
        # Getting the test
        elif line.lstrip().startswith('Test:'):
            test_divisor = int(line.split('by')[1].lstrip())
            
        # Getting the condition when the test is true
        elif line.lstrip().startswith('If true:'):
            true_condition = int(line.split('monkey')[1].lstrip())
            
        # Getting the condition when the test is false
        elif line.lstrip().startswith('If false:'):
            false_condition = int(line.split('monkey')[1].lstrip())
            
    # Joining all the details in a single dictionary
    monkey_details = {
        'currently_holding': currently_holding,
        'operations': operations,
        'test_divisor': test_divisor,
        'true_condition': true_condition,
        'false_condition': false_condition,
        'times_inspected': 0
    }
        
    return monkey_number, monkey_details
````

````python
# Loading in the monkeys from the input file
# with open('dummy.txt', 'r') as f:    
with open('input.txt', 'r') as f:
    monkey_input = f.read().splitlines()
    monkey_input.append('')

# Instantiating an empty dictionary to hold the structured monkey information
monkey_dict = {}

# Instantiating an empty list to hold each individual monkey's information
monkey_info = []

# Iterating over all the inputs
for line in monkey_input:
    
    # Appending the line for each monkey detail
    monkey_info.append(line)
    
    # Parsing through the information once collected for an individual monkey
    if line == '':
        
        # Extracting the monkey number and details
        monkey_number, monkey_details = parse_monkey_info(monkey_info)
        
        # Appending the monkey number and dictionary to the collective list
        monkey_dict[monkey_number] = monkey_details
        
        # Reinstantiating the monkey list for the next line
        monkey_info = []
````

````python
# Continuing the process for X number of rounds
rounds = 20
for _ in range(rounds):

    # Iterating over each monkey
    for monkey in monkey_dict.keys():
        
        # Iterating over any item that the monkey is currently holding
        while len(monkey_dict[monkey]['currently_holding']) > 0:
            
            # Popping the first item the monkey is holding
            held_item = monkey_dict[monkey]['currently_holding'].pop(0)
            
            # Increasing the number of times a monkey has inspected an item
            monkey_dict[monkey]['times_inspected'] += 1
            
            # Extracting the math operator
            math_operator = monkey_dict[monkey]['operations'][0]
            
            # Extracting the number alongside the math operator
            math_operator_number = monkey_dict[monkey]['operations'][1]
            
            # Converting any "old" to the number of the held item
            if math_operator_number == 'old':
                math_operator_number = str(held_item)
            
            # Updating the worry level of the item
            new_worry_item = math.floor(eval(f'{held_item} {math_operator} {math_operator_number}') / 3)
            
            # Checking to see if the new worry item passes or fails their internal test
            if new_worry_item % monkey_dict[monkey]['test_divisor'] == 0:
                
                # Passing the item per the "True" test case
                passing_to = monkey_dict[monkey]['true_condition']
                monkey_dict[passing_to]['currently_holding'].append(new_worry_item)
                
            else:
                
                # Passing the item per the "False" test case
                passing_to = monkey_dict[monkey]['false_condition']
                monkey_dict[passing_to]['currently_holding'].append(new_worry_item)
            
            
````

````python
# Getting a sorted list of all times inspected
all_times_inspected = sorted([monkey[1]['times_inspected'] for monkey in monkey_dict.items()], reverse = True)

# Getting the final result
final_result = all_times_inspected[0] * all_times_inspected[1]
print(f'Final result: {final_result}')
````
