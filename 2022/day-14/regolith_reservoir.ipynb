Sand is produced one unit at a time, and the next unit of sand is not produced until the previous unit of sand comes to rest. A unit of sand is large enough to fill one tile of air in your scan.

A unit of sand always falls down one step if possible. If the tile immediately below is blocked (by rock or sand), the unit of sand attempts to instead move diagonally one step down and to the left. If that tile is blocked, the unit of sand attempts to instead move diagonally one step down and to the right. Sand keeps moving as long as it is able to do so, at each step trying to move down, then down-left, then down-right. If all three possible destinations are blocked, the unit of sand comes to rest and no longer moves, at which point the next unit of sand is created back at the source.

````python
# Loading the input file
with open('input.txt', 'r') as f:
    rock_lines = f.read().splitlines()
    
dummy_input = [
    '498,4 -> 498,6 -> 496,6',
    '503,4 -> 502,4 -> 502,9 -> 494,9'
]

# rock_lines = dummy_input
````

````python
def parse_rock_lines(rock_lines):
    '''
    Parses through the lines of rocks
    
    Inputs:
        - rock_lines (list): List of strings representing the rock lines
        
    Returns:
        - spaces_filled (set): A set containing all the spaces filled in by the rocks
    '''
    
    # Instantiating a set to add filled in spaces
    spaces_filled = set()
    
    # Iterating over each rock line
    for rock_line in rock_lines:
        
        # Instantiating a container to hold all the "corner points"
        corner_points = []
        
        # Splitting the raw corners based on the "arrow" characters
        raw_corners = rock_line.split(' -> ')
        
        # Iterating over the raw corners
        for raw_corner in raw_corners:
            
            # Extracting the x- and y- coordinate positions as integers
            x = int(raw_corner.split(',')[0])
            y = int(raw_corner.split(',')[1])
            
            # Appending the corner as a new tuple
            corner_points.append((x, y))
            
        while len(corner_points) > 1:
            
            # Getting the starting / ending corner points for each line segment
            starting_corner = corner_points.pop(0)
            ending_corner = corner_points[0]
            
            # Checking if we are drawing a vertical or horizontal line
            if starting_corner[0] == ending_corner[0]:
                
                # Fixing the x position
                x = starting_corner[0]
                
                # Setting the directionality of the line
                if starting_corner[1] < ending_corner[1]:
                    start = starting_corner[1]
                    finish = ending_corner[1]
                elif starting_corner[1] > ending_corner[1]:
                    start = ending_corner[1]
                    finish = starting_corner[1]
                
                # Adding the points of the horizontal line
                for y in range(start, finish + 1):
                    spaces_filled.add((x, y))
                    
            elif starting_corner[1] == ending_corner[1]:
                
                # Fixing the y position
                y = starting_corner[1]
                
                # Setting the directionality of the line
                if starting_corner[0] < ending_corner[0]:
                    start = starting_corner[0]
                    finish = ending_corner[0]
                elif starting_corner[0] > ending_corner[0]:
                    start = ending_corner[0]
                    finish = starting_corner[0]
                
                # Adding the points of the vertical line
                for x in range(start, finish + 1):
                    spaces_filled.add((x, y))
    
    return spaces_filled
````

````python
spaces_filled = parse_rock_lines(rock_lines)
````

````python
# Setting the starting position of the sandhole
sandhole = (500, 0)

# Getting the abyss position
abyss = max([space_filled[1] for space_filled in spaces_filled]) + 1

# Instantiating number of sand falling
grains_of_sand = 1

# Starting the current grain of sand from the sandhole
current_position = sandhole

while True:
    # Checking if the sand is falling into the abyss
    if current_position[1] == abyss:
        print(f'Grains of sand: {grains_of_sand - 1}')
        break

    # Checking if the sand can head straight down
    if (current_position[0], current_position[1] + 1) not in spaces_filled:
        current_position = (current_position[0], current_position[1] + 1)

    # Checking if the sand can move down and toward the left
    elif (current_position[0] - 1, current_position[1] + 1) not in spaces_filled:
        current_position = (current_position[0] - 1, current_position[1] + 1)
        
    # Checking if the sand can move down and toward the right
    elif (current_position[0] + 1, current_position[1] + 1) not in spaces_filled:
        current_position = (current_position[0] + 1, current_position[1] + 1)

    # Filling in any space if the sand can't keep moving
    else:       
        spaces_filled.add(current_position)
    
        # Resetting the position for the next grain of sand
        current_position = sandhole
        grains_of_sand += 1
````

````python
spaces_filled = parse_rock_lines(rock_lines)

# Setting the starting position of the sandhole
sandhole = (500, 0)

# Getting the floor position
floor = max([space_filled[1] for space_filled in spaces_filled]) + 2

# Instantiating number of sand falling
grains_of_sand = 1

# Starting the current grain of sand from the sandhole
current_position = sandhole

while True:
    
    # Checking if the sand has reached the top
    if ((501, 1) in spaces_filled):
        print(f'Grains of sand: {grains_of_sand}')
        break
        
    # Checking if the sand has hit the floor
    elif current_position[1] == floor:
        grains_of_sand += 1
        spaces_filled.add((current_position[0], current_position[1] - 1))
        current_position = sandhole
        
    # Checking if the sand can head straight down
    elif (current_position[0], current_position[1] + 1) not in spaces_filled:
        current_position = (current_position[0], current_position[1] + 1)

    # Checking if the sand can move down and toward the left
    elif (current_position[0] - 1, current_position[1] + 1) not in spaces_filled:
        current_position = (current_position[0] - 1, current_position[1] + 1)
        
    # Checking if the sand can move down and toward the right
    elif (current_position[0] + 1, current_position[1] + 1) not in spaces_filled:
        current_position = (current_position[0] + 1, current_position[1] + 1)     
    
    # Starting the next grain of sand
    else:
        spaces_filled.add(current_position)
    
        # Resetting the position for the next grain of sand
        current_position = sandhole
        grains_of_sand += 1
````
