# Day 2 - Rock Paper Scissors

## Part 1

The Elves begin to set up camp on the beach. To decide whose tent gets to be closest to the snack storage, a giant Rock Paper Scissors tournament is already in progress.

Rock Paper Scissors is a game between two players. Each game contains many rounds; in each round, the players each simultaneously choose one of Rock, Paper, or Scissors using a hand shape. Then, a winner for that round is selected: Rock defeats Scissors, Scissors defeats Paper, and Paper defeats Rock. If both players choose the same shape, the round instead ends in a draw.

Appreciative of your help yesterday, one Elf gives you an encrypted strategy guide (your puzzle input) that they say will be sure to help you win. "The first column is what your opponent is going to play: A for Rock, B for Paper, and C for Scissors. The second column--" Suddenly, the Elf is called away to help with someone's tent.

The second column, you reason, must be what you should play in response: X for Rock, Y for Paper, and Z for Scissors. Winning every time would be suspicious, so the responses must have been carefully chosen.

The winner of the whole tournament is the player with the highest score. Your total score is the sum of your scores for each round. The score for a single round is the score for the shape you selected (1 for Rock, 2 for Paper, and 3 for Scissors) plus the score for the outcome of the round (0 if you lost, 3 if the round was a draw, and 6 if you won).

Since you can't be sure if the Elf is trying to help you or trick you, you should calculate the score you would get if you were to follow the strategy guide.

For example, suppose you were given the following strategy guide:

```
A Y
B X
C Z
```

This strategy guide predicts and recommends the following:

In the first round, your opponent will choose Rock (A), and you should choose Paper (Y). This ends in a win for you with a score of 8 (2 because you chose Paper + 6 because you won).
In the second round, your opponent will choose Paper (B), and you should choose Rock (X). This ends in a loss for you with a score of 1 (1 + 0).
The third round is a draw with both players choosing Scissors, giving you a score of 3 + 3 = 6.
In this example, if you were to follow the strategy guide, you would get a total score of 15 (8 + 1 + 6).

What would your total score be if everything goes exactly according to your strategy guide?

````python
# Importing the rock paper scissors (RPS) games from a text file
with open('input.txt', 'r') as f:
    rps_games = f.read().splitlines()
````

````python
# Creating a custom function to remap the input values from ABC / XYZ to rock, paper, scissors
def remap_rps_input(current_mapping):
    '''
    Remaps the input values from ABC / XYZ to a more understandable rock, paper, scissors
    
    Inputs:
        - current_mapping (str): The current input choice provided by the player in ABC / XYZ form
    
    Returns:
        - rps_choice (str): A more understandable rock, paper, scissors choice
    '''
    
    # If A or X, remapping to "rock"
    if current_mapping == 'A' or current_mapping == 'X':
        rps_choice = 'rock'
    # If B or Y, remapping to "paper"
    elif current_mapping == 'B' or current_mapping == 'Y':
        rps_choice = 'paper'
    # If C or Z, remapping to "scissors"
    elif current_mapping == 'C' or current_mapping == 'Z':
        rps_choice = 'scissors'
        
    return rps_choice
````

````python
# Creating a custom function to get the game score for each game in the full set of games provided
def get_game_score(game):
    '''
    Determines the proper score of the Rock Paper Scissors game
    
    Inputs:
        - game (str): A string value containing the choices of both players
        
    Returns:
        - game_score (int): An integer representing the score of this particular round of Rock Paper Scissors
    '''
    
    # Instantiating a score for this particular game
    game_score = 0
    
    # Parsing the player entries from the game and remapping it on the fly
    other_player = remap_rps_input(game.split()[0])
    your_choice = remap_rps_input(game.split()[1])
    
    # Determining the initial score value per "my" choice
    if your_choice == 'rock':
        game_score += 1
    elif your_choice == 'paper':
        game_score += 2
    elif your_choice == 'scissors':
        game_score += 3

    # Adding 6 points for each scenario in which you win
    if your_choice == 'rock' and other_player == 'scissors':
        game_score += 6
    elif your_choice == 'paper' and other_player == 'rock':
        game_score += 6
    elif your_choice == 'scissors' and other_player == 'paper':
        game_score += 6
    # Adding 3 points in the case that you draw
    elif your_choice == other_player:
        game_score += 3
        
    return game_score
````

````python
# Instantiating a variable for cummulative score
cumm_score = 0
        
# Iterating through all the RPS games
for game in rps_games:
    
    # Adding the score of the individual game to the cummulative_score
    cumm_score += get_game_score(game)
    
# Printing the final results
print(f'Your total score after all the Rock Paper Scissors games: {cumm_score}')
````

## Part 2

The Elf finishes helping with the tent and sneaks back over to you. "Anyway, the second column says how the round needs to end: X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win. Good luck!"

The total score is still calculated in the same way, but now you need to figure out what shape to choose so the round ends as indicated. The example above now goes like this:

- In the first round, your opponent will choose Rock (A), and you need the round to end in a draw (Y), so you also choose Rock. This gives you a score of 1 + 3 = 4.
- In the second round, your opponent will choose Paper (B), and you choose Rock so you lose (X) with a score of 1 + 0 = 1.
- In the third round, you will defeat your opponent's Scissors with Rock for a score of 1 + 6 = 7.
Now that you're correctly decrypting the ultra top secret strategy guide, you would get a total score of 12.

Following the Elf's instructions for the second column, what would your total score be if everything goes exactly according to your strategy guide?

````python
# Creating a custom function that determines the strategic choice you'll have to make per the elf's input
def get_strategic_choice_points(strategy, other_player):
    '''
    Determines how many points to add per the strategic choice recommended by the elf
    
    Inputs:
        - strategy (str): The strategy recommended by the elf ("X" is lose, "Y" is draw, "Z" is win)
        - other_player (str): The current choice made by the other player
        
    Returns:
        - score (int): The number of points that will need added to the RPS game
    '''
    
    # Instantiating a value determining the score we'll be returning from this function
    score = 0
    
    # Determining what to pick if the elf recommends that you lose
    if strategy == 'X':
        
        # If rock, adding 3 points as you'll have to pick scissors
        if other_player == 'rock':
            score += 3
            
        # If paper, adding 1 point as you'll have to pick rock
        elif other_player == 'paper':
            score += 1
        
        # If scissors, adding 2 points as you'll have to pick paper
        elif other_player == 'scissors':
            score += 2
            
    # Determining what to pick if the elf recommends that you tie
    elif strategy == 'Y':
        
        # If rock, adding 1 point as you'll have to pick rock too
        if other_player == 'rock':
            score += 1
            
        # If paper, adding 2 points as you'll have to pick paper too
        elif other_player == 'paper':
            score += 2
        
        # If scissors, adding 3 points as you'll have to pick scissors too
        elif other_player == 'scissors':
            score += 3
            
    # Determining what to pick if the elf recommends that you win
    elif strategy == 'Z':
        
        # If rock, adding 2 points as you'll have to pick paper
        if other_player == 'rock':
            score += 2
            
        # If paper, adding 3 points as you'll have to pick scissors
        elif other_player == 'paper':
            score += 3
        
        # If scissors, adding 1 point as you'll have to pick rock
        elif other_player == 'scissors':
            score += 1
            
    return score
````

````python
# Creating a custom function to get the game score with the elf's new instructions
def get_elf_game_score(game):
    '''
    Determines the proper score of the Rock Paper Scissors game
    
    Inputs:
        - game (str): The choice of the other player and the elf's indication on what you should do
        
    Returns:
        - game_score (int): The score of this particular round of Rock Paper Scissors
    '''
    
    # Instantiating a score for this particular game
    game_score = 0
    
    # Parsing out the other player's choice and the elf's recommended strategy
    other_player = remap_rps_input(game.split()[0])
    strategy = game.split()[1]
    
    # Adding points you'll receive per the elf's recommended strategy
    if strategy == 'Y':
        game_score += 3
    elif strategy == 'Z':
        game_score += 6
            
    # Adding the number of points given the choice you'll have to make per what the elf recommends
    game_score += get_strategic_choice_points(strategy, other_player)
    
    return game_score
````

````python
# Instantiating a variable for cummulative score
cumm_score = 0
        
# Iterating through all the RPS games
for game in rps_games:
    
    # Adding the score of the individual game to the cummulative_score
    cumm_score += get_elf_game_score(game)
    
# Printing the final results
print(f'Your total score after all the Rock Paper Scissors games with the Elf\'s new rules: {cumm_score}')
````
